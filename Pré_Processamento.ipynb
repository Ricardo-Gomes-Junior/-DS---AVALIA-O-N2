```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Estar lendo dados de arquivos CSV de 2013 até 2024.
df2024 = pd.read_csv('/content/datatran2024.csv', encoding='latin-1', delimiter=';')
df2023 = pd.read_csv('/content/datatran2023.csv', encoding='latin-1', delimiter=';')
df2022 = pd.read_csv('/content/datatran2022.csv', encoding='latin-1', delimiter=';')
df2021 = pd.read_csv('/content/datatran2021.csv', encoding='latin-1', delimiter=';')
df2020 = pd.read_csv('/content/datatran2020.csv', encoding='latin-1', delimiter=';')
df2019 = pd.read_csv('/content/datatran2019.csv', encoding='latin-1', delimiter=';')
df2018 = pd.read_csv('/content/datatran2018.csv', encoding='latin-1', delimiter=';')
df2017 = pd.read_csv('/content/datatran2017.csv', encoding='latin-1', delimiter=';')
df2016 = pd.read_csv('/content/datatran2016.csv', encoding='latin-1', delimiter=';')
df2015 = pd.read_csv('/content/datatran2015.csv', encoding='latin-1', delimiter=';')
df2014 = pd.read_csv('/content/datatran2014.csv', encoding='latin-1', delimiter=';')
df2013 = pd.read_csv('/content/datatran2013.csv', encoding='latin-1', delimiter=';')
```

```python
# Concatenando os DataFrames um único conjunto e exibindo o número de linhas e colunas
result = pd.concat([df2024, df2023, df2022, df2021, df2020, df2019, df2018, df2017, df2016, df2015, df2014, df2013])
result.head()
```

```python
# Exibir o número de linhas e colunas do DataFrame result
num_linhas, num_colunas = result.shape
print("Número de linhas:", num_linhas)
print("Número de colunas:", num_colunas)
```

```python
# Exibindo a quantidade de valores nulos nas colunas.
nulls_columns = result.isnull().sum()
print (nulls_columns)
```

```python
# Apaga colunas desnecessárias e colunas com valores nulos.
df = result.drop(columns=['sentido_via','uso_solo','regional','delegacia','uop', 'ignorados', 'latitude', 'longitude', 'ano'])
```

```python
# Filtrar os dados dos estados do nordeste
estados_nordeste = ['PE', 'PB', 'AL', 'SE', 'BA', 'CE', 'MA', 'PI', 'RN']

# Use a função isin() para selecionar apenas as linhas onde o estado ('uf') está na lista de estados do Nordeste
df_nordeste = df[df['uf'].isin(estados_nordeste)]
```

```python
df_nordeste.head()
```

```python
df_nordeste.to_csv('df_nordeste.csv', index=False)
```

# Números de Acidentes por Hora do Dia

```python
# Carregando os dados
df = pd.read_csv('/content/df_nordeste.csv', parse_dates=['data_inversa'], dayfirst=True)

# Extraindo a hora do dia
df['hora'] = df['horario'].str.split(':').str[0]
df['hora'] = df['hora'].astype(int)

# Contagem de acidentes por hora do dia
acidentes_por_hora = df.groupby('hora').size()
```

```python
# Visualizando a série temporal
plt.figure(figsize=(10, 6))
acidentes_por_hora.plot(marker='o')
plt.title('Número de Acidentes de Trânsito por Hora do Dia (Nordeste)')
plt.xlabel('Hora do Dia')
plt.ylabel('Número de Acidentes')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```
